```python
import pandas as pd
import numpy as np
import tensorflow as tf
import re
```

```python
games_df = pd.read_csv("../Resources/vgsales.csv.zip")
ratings_df = pd.read_csv("../Resources/metacritic_games.csv")
```

```python
ratings_df.rename({"game":"Name"},axis=1, inplace=True)
ratings_df
```

```python
combo_df = pd.merge(games_df, ratings_df, on="Name", how="inner")
```

```python
# Not enough crossover
combo_df
```

```python
games_list = games_df["Name"].values.tolist()
ratings_list = ratings_df["Name"].values.tolist()
print(len(games_list))
print(len(ratings_list))
```

```python
g_standardized_titles=[]
for game in games_list:
    x = re.compile("\W|\s")
    edited_title= re.sub(x,"-", game).lower().replace("--","-")
    if edited_title[-1] == "-":
        edited_title = edited_title[:-1]
    g_standardized_titles.append(edited_title)
```

```python
r_standardized_titles=[]
for game in ratings_list:
    x = re.compile("\W|\s")
    edited_title= re.sub(x,"-", game).lower().replace("--","-")
    if edited_title[-1] == "-":
        edited_title = edited_title[:-1]
    r_standardized_titles.append(edited_title)
```

```python
print(len(g_standardized_titles))
print(len(r_standardized_titles))
```

```python
for game in g_standardized_titles:
    if game in r_standardized_titles:
        g_standardized_titles.remove(game)
        r_standardized_titles.remove(game)
    else:
        continue
```
